```python
import numpy as np
import pandas as pd
import os
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#read file

covid19India_df = pd.read_csv('covid_19_india.csv')
covid19India_df
```

```python
covid19India_df.shape
```

```python
covid19India_df.info()
```

```python
covid19India_df.head()
```

```python
# Set the width and height of the figure
plt.figure(figsize=(14,6))

# Add title
plt.title("Full dataset")

# Line chart  
sns.lineplot(data=covid19India_df)
```

```python
# State wise Observations
covid19India_df['State/UnionTerritory'].value_counts().sort_values() 
```

```python
#higher number of entries
covid19India_df_Haryana = covid19India_df[covid19India_df['State/UnionTerritory']=='Haryana']
covid19India_df_Haryana
covid19India_df_Kerala = covid19India_df[covid19India_df['State/UnionTerritory']=='Kerala']
covid19India_df_Kerala
covid19India_df_WestBengal = covid19India_df[covid19India_df['State/UnionTerritory']=='West Bengal']
covid19India_df_WestBengal
```

```python
plt.figure(figsize=(10,6))
plt.title("Haryana dataset")
sns.lineplot(data=covid19India_df_Haryana)
```

```python
plt.figure(figsize=(10,6))
plt.title("Kerala dataset")
sns.lineplot(data=covid19India_df_Kerala)
```

```python
plt.figure(figsize=(10,6))
plt.title("WestBengal dataset")
sns.lineplot(data=covid19India_df_WestBengal)
```

```python
covid19India_df_Telengana = covid19India_df[covid19India_df['State/UnionTerritory']=='Telengana']
covid19India_df_Telengana
plt.figure(figsize=(10,6))
plt.title(" dataset Telengana")
sns.lineplot(data=covid19India_df_Telengana)
```

```python
covid19India_df_by_State = covid19India_df.groupby('State/UnionTerritory')[['Cured','Deaths' ,'Confirmed']].sum().reset_index()
covid19India_df_by_State
```

```python
plt.figure(figsize=(10,6))
plt.title("Cured")
sns.barplot(x=covid19India_df_by_State.index, y=covid19India_df_by_State['Cured'])
plt.ylabel("Cured")
```

```python
plt.figure(figsize=(10,6))
plt.title("Confirmed")
sns.barplot(x=covid19India_df_by_State.index, y=covid19India_df_by_State['Confirmed'])
plt.ylabel("Confirmed")
```

```python
plt.figure(figsize=(10,6))
plt.title("Deaths")
sns.barplot(x=covid19India_df_by_State.index, y=covid19India_df_by_State['Deaths'])
plt.ylabel("Deaths")
```
